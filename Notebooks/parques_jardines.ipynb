```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Raw_Datasets/Trabajado/200761-0-parques-jardines.csv',sep=';',encoding='latin-1')
df
```

```python
df.info()
```

```python
# Eliminar columnas con muchos valores nulos
df = df.drop(columns=['PLANTA', 'ESCALERAS', 'Unnamed: 32', 'PUERTA','ORIENTACION'])
```

```python
df = df.drop(columns=['FAX','TELEFONO','TIPO','EMAIL','DESCRIPCION', 'DESCRIPCION-ENTIDAD'])
```

```python
mean_num = df['NUM'].mean()
df['NUM'].fillna(mean_num, inplace=True)
```

```python
df['EQUIPAMIENTO'].fillna('No especificado', inplace=True)
```

```python
df.info()
```

```python
df
```

```python
valores_nulos_por_distrito = df.groupby('BARRIO')['DISTRITO'].apply(lambda x: x.isnull().sum()).reset_index()
```

```python
valores_nulos_por_distrito.rename(columns={'DISTRITO': 'VALORES_NULOS'}, inplace=True)
```

```python
print(valores_nulos_por_distrito)
```

```python
valores_nulos_por_distrito = df.groupby('DISTRITO')['CODIGO-POSTAL'].apply(lambda x: x.isnull().sum()).reset_index()
```

```python
valores_nulos_por_distrito.rename(columns={'CODIGO-POSTAL': 'VALORES_NULOS'}, inplace=True)
```

```python
# Mostrar el resultado
print(valores_nulos_por_distrito)
```

```python

# Creamos un diccionario con los valores que tienes disponibles para cada distrito
valores_distritos = {
   
    'MONCLOA-ARAVACA': 28082,
     'DISTRITO':28005
    # Agrega los demás distritos y sus valores correspondientes aquí
}

# Rellenar los valores nulos en la columna "COD-DISTRITO" utilizando el diccionario
df['CODIGO-POSTAL'] = df['CODIGO-POSTAL'].fillna(df['DISTRITO'].map(valores_distritos))

# Mostrar el DataFrame resultante
print(df)
```
