# Limpieza y Regeneración de Datos

```python
clima = [col for col in dep.columns if col.startswith('Clima_')]
dep.drop(columns=clima, inplace=True)
dep
```

```python
condicion = [col for col in dep.columns if col.startswith('Condicion_')]
dep['Condición'] = dep[condicion].idxmax(axis=1).str.replace('Condicion_', '')
dep.drop(columns=condicion, inplace=True)
dep
```

```python
genero = [col for col in dep.columns if col.startswith('Genero_')]
dep['Género'] = dep[genero].idxmax(axis=1).str.replace('Genero_', '')
dep.drop(columns=genero, inplace=True)
dep
```

```python
objetivo = [col for col in dep.columns if col.startswith('Objetivo_')]
dep['Objetivo'] = dep[objetivo].idxmax(axis=1).str.replace('Objetivo_', '')
dep.drop(columns=objetivo, inplace=True)
sports_columns = dep.columns[5:61]
dep.drop(columns=sports_columns, inplace=True)
dep
```

```python
dep.rename(columns={'Objetivo' : 'Nivel de Objetivo Calórico'}, inplace=True)
dep
```

```python
probabilidades_dict = {'Aeróbicos': 8, 'Aeróbicos acuáticos': 5, 'Artes marciales': 7, 'Atletismo': 9, 'BMX': 2,
                       'Baloncesto': 10,'Balonmano': 6,'Billar': 3,'Bolos': 3,'Boxeo': 5,'Bádminton': 6,'Béisbol': 8,
                       'Calistenia': 4,'Calva': 1,'Caminar': 7,'Chito': 1,'Ciclismo': 9,'Ciclismo estacionario': 3,
                       'Correr': 10,'Dardos': 5,'Entrenamiento en circuito': 7,'Escalada': 6,'Frisbee': 3,'Frontenis': 2,'Fútbol': 10,
                       'Fútbol sala': 6,'Gimnasia': 4,'Golf': 6,'Hockey': 8,'Kickball': 2,'Kickboxing': 3,'Levantamiento de peso': 4,
                       'Marcha rápida': 8,'Minigolf': 2,'Montañismo': 5,'Máquina escaladora': 3,'Nado sincronizado': 2,'Natación': 9,
                       'Padel': 10,'Patinaje': 5,'Patinaje sobre hielo': 3,'Petanca': 1,'Raquetbol': 2,'Salto a la comba': 1,
                       'Senderismo': 3,'Skateboard': 3,'Sóftbol': 4,'Tai chi': 2,'Tenis': 8,'Tenis de mesa': 5,'Tenis en pareja': 3,
                       'Ultimate frisbee': 3,'Voleibol': 7,'Voleibol acuático': 4,'Waterpolo': 3,'Yoga': 2}

# Normalizar las probabilidades para que sumen 1
total_probabilidades = sum(probabilidades_dict.values())
probabilidades_dict = {deporte: probabilidad / total_probabilidades for deporte, probabilidad in probabilidades_dict.items()}

# Crear una función para seleccionar deportes con sesgo aleatoriamente
def deporte_favorito():
    return random.choices(list(probabilidades_dict.keys()), weights=list(probabilidades_dict.values()))[0]

# Agregar la columna 'Deporte Favorito' utilizando la función para seleccionar deportes aleatoriamente
dep['Deporte Favorito'] = dep.apply(lambda row: deporte_favorito(), axis=1)
dep
```

```python
def generar_temperatura_madrid():
    temperatura_minima = 19 
    temperatura_media = 32  
    sesgo = -1.2     
    temperatura_aleatoria = random.gauss(temperatura_media, sesgo)

    temperatura_final = max(temperatura_minima, temperatura_aleatoria)
    
    return temperatura_final
dep['Temperatura (°C)'] = dep.apply(lambda row: generar_temperatura_madrid(), axis=1)
dep['Temperatura (°C)'] = dep['Temperatura (°C)'].round(2)
```

```python
def generar_humedad_madrid():
    humedad_minima = 27 
    humedad_media = 40 
    sesgo = 0.8
    
    humedad_aleatoria = random.gauss(humedad_media, sesgo)
    
    humedad_final = max(humedad_minima, humedad_aleatoria)
    
    return humedad_final

dep['Humedad'] = dep.apply(lambda row: generar_humedad_madrid(), axis=1)
dep['Humedad'] = (dep['Humedad']).round(2)
```

```python
def generar_genero():
    opciones = ["Hombre", "Mujer", "Otro"]  
    probabilidades = [0.435, 0.535, 0.03] 
    return random.choices(opciones, weights=probabilidades)[0]

def generar_altura(genero):
    if genero == "Hombre":
        altura_media = 1.76
        desviacion_estandar = 0.07
    elif genero == "Mujer":
        altura_media = 1.62
        desviacion_estandar = 0.06
    else:
        altura_minima = 1.65
        altura_maxima = 1.77
        return random.uniform(altura_minima, altura_maxima)

    altura_aleatoria = random.gauss(altura_media, desviacion_estandar)
    return max(1.0, altura_aleatoria)

def generar_peso(genero):
    if genero == "Hombre":
        peso_media = 84.0
        desviacion_estandar = 6.0
    elif genero == "Mujer":
        peso_media = 65.9
        desviacion_estandar = 5.0
    else:
        peso_minimo = 68.0
        peso_maximo = 100.0
        return random.uniform(peso_minimo, peso_maximo)

    peso_aleatorio = random.gauss(peso_media, desviacion_estandar)
    return max(30.0, peso_aleatorio)

def calcular_imc(altura, peso):
    return peso / altura ** 2

dep['Género'] = dep.apply(lambda row: generar_genero(), axis=1)
dep['Altura (m)'] = dep.apply(lambda row: generar_altura(row['Género']), axis=1)
dep['Peso (Kg)'] = dep.apply(lambda row: generar_peso(row['Género']), axis=1)
```

```python
dep[['Género', 'Altura (m)', 'Peso (Kg)', ]] = (dep[['Género', 'Altura (m)', 'Peso (Kg)']]).round(2)
dep['IMC'] = dep.apply(lambda row: calcular_imc(row['Altura (m)'], row['Peso (Kg)']), axis=1)
dep.IMC = dep.IMC.round(2)
```

```python
def clasificar_actividad_aleatoria(imc):
    prob_actividad = random.uniform(0, 1)
    
    if imc < 18.5:
        if prob_actividad < 0.7: 
            return "Alta"
        else:
            return "Baja"
    elif imc >= 18.5 and imc <= 24.9:
        if prob_actividad < 0.5:  
            return "Alta"
        else:
            return "Moderada"
    else:
        if prob_actividad < 0.3: 
            return "Alta"
        else:
            return "Baja"

dep['Actividad'] = dep['IMC'].apply(clasificar_actividad_aleatoria)
dep
```

```python
dep.drop(columns='Condición', inplace=True)
dep[dep['Actividad'] == 'Moderada']
```

```python
def clasificar_clima_aleatorio():
    opciones = ["Soleado", "Nublado", "Lluvioso"]  # Opciones de clima
    probabilidades = [0.8, 0.1, 0.1]  # Probabilidades para cada opción (Soleado, Nublado, Lluvioso)
    return random.choices(opciones, weights=probabilidades)[0]

# Agregar la columna 'Clima' utilizando la función para clasificar el clima aleatoriamente
dep['Clima'] = dep.apply(lambda row: clasificar_clima_aleatorio(), axis=1)
```
