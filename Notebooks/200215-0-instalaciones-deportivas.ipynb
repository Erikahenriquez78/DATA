```python
import re

nombres_deportes = ['bici-trial', 'mini pitch', 'calva', 'padel', 'rocódromo', 'tenis', 'tenis de mesa', 'baloncesto', 'fútbol sala', 'fútbol', 'balonmano', 'minibaloncesto', 'minifútbol', 'voleibol', 'skate', 'pista de patinaje', 'automodelismo', 'hockey', 'bicicletas', 'skateboard', 'hockey patines', 'pádel', 'parkour', 'calistenia', 'footing', 'bádminton', 'bolos', 'petanca', 'chito', 'fútbol 7', 'patinódromo', 'hockey sobre patines', 'frontón', 'fútbol-sala', 'bolos y petanca', 'patinaje y hockey sobre patines', 'vóley y futvoley', 'atletismo', 'rugby']
# Función para extraer deportes de una frase
def extraer_deportes(frase):

    # Convertir la frase a minúsculas para una búsqueda sin distinción de mayúsculas y minúsculas
    frase = frase.lower()

    deportes_encontrados = []
    for deporte in nombres_deportes:
        # Buscar el nombre del deporte en la frase, asegurándose de que esté precedido y seguido por un espacio o límites de palabra
        if re.search(r'\b' + re.escape(deporte) + r'\b', frase):
            deportes_encontrados.append(deporte)
    return deportes_encontrados

# Lista de instalaciones que contiene deportes
instalaciones = data['EQUIPAMIENTO'].dropna().unique()

# Conjunto para almacenar los deportes únicos encontrados en los elementos de la lista
deportes_unicos = set()

# Recorrer cada elemento de la lista y extraer los deportes
for elemento in instalaciones:
    deportes = extraer_deportes(elemento)
    for deporte in deportes:
        # Convertir a minúsculas y agregar al conjunto de deportes únicos
        deportes_unicos.add(deporte.lower())

# Lista de deportes en la lista nombres_deportes que no han sido encontrados
deportes_no_encontrados = [deporte for deporte in nombres_deportes if deporte.lower() not in deportes_unicos]

# Imprimir la lista de deportes únicos encontrados en los elementos de la lista
print("Deportes únicos encontrados:", list(deportes_unicos))

# Imprimir la lista de deportes no encontrados
print("Deportes no encontrados:", deportes_no_encontrados)

# Realizar una diferencia entre la lista de instalaciones y los deportes encontrados
instalaciones_sin_deportes = [elemento for elemento in instalaciones if not any(deporte.lower() in elemento.lower() for deporte in list(deportes_unicos))]

# Imprimir la lista de instalaciones sin los deportes encontrados
print("Lista de instalaciones sin deportes extraídos:")
for elemento in instalaciones_sin_deportes:
    print(elemento)
```
