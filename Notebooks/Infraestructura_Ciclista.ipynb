```python
import pandas as pd
from shapely.wkt import loads
import utm

# Supongamos que data['geometry'] contiene los datos con la coordenada Z
data = pd.DataFrame({'geometry': ['LINESTRING Z (437719.5965 4476171.279200001 0, 437724.90029999986 4476166.3467 0, 437763.01410000026 4476117.3904 0)',
                                  'LINESTRING Z (438719.5965 4476171.279200001 0, 438724.90029999986 4476166.3467 0, 438763.01410000026 4476117.3904 0)']})

# Define una función para convertir geometrías WKT a objetos Shapely
def wkt_to_geom(wkt):
    return loads(wkt)

# Convierte la columna 'geometry' del DataFrame a geometrías de Shapely
data['geometry'] = data['geometry'].apply(wkt_to_geom)

# Define una función para eliminar la coordenada Z (elevación) de una geometría
def remove_z_coord(geom):
    if geom.has_z:
        if isinstance(geom, LineString):
            return LineString([(x, y) for x, y, _ in geom.coords])
        elif isinstance(geom, MultiLineString):
            return MultiLineString([[xy[:2] for xy in part.coords] for part in geom.geoms])
    return geom

# Elimina la coordenada Z (elevación) de cada geometría
data['geometry'] = data['geometry'].apply(remove_z_coord)

# Define una función para convertir las coordenadas UTM a latitud y longitud
def utm_to_latlon(row):
    coords_list = list(row['geometry'].coords)
    return [utm.to_latlon(easting, northing, 30, "T")[:2] for easting, northing, _ in coords_list]

# Convierte las coordenadas UTM a latitud y longitud y crea una nueva columna 'lat_lon_list'
data['lat_lon_list'] = data.apply(utm_to_latlon, axis=1)

# Ahora data['lat_lon_list'] contiene las coordenadas en formato (latitud, longitud) sin la elevación
print(data)
```
